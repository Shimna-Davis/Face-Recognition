```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('/home/ann/computervision/opencv/bla2.jpg')
```

```python
facecascade=cv2.CascadeClassifier('/home/ann/computervision/opencv/facedetection/haarcascade_frontalface_default.xml')
```

```python
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(converted_img)
```

```python
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
faces=facecascade.detectMultiScale(img_grey,scaleFactor=1.1,minNeighbors=4,minSize=(30,30))
```

```python
faces
```

```python
len(faces)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),(0,255,0),2)
```
